```python
import pandas as pd

df_train = pd.read_csv("Data/GUIDE_train.csv", nrows=100000)
cols = list(df_train.columns)
```

```python
df_train
```

```python
Id_columns = ['Id', 'OrgId', 'IncidentId', 'AlertId']

cols = [col for col in cols if col not in Id_columns]
len(cols)
```

```python
pct40_blank_cols = []
pct50_blank_cols = []
pct60_blank_cols = []
pct70_blank_cols = []
pct80_blank_cols = []
pct90_blank_cols = []


for col in cols:
    df = pd.read_csv("Data/GUIDE_train.csv", usecols=[col])
    total_cols = df.shape[0]
    blank_cols_no =  df.isna().sum()[list((df.isna().sum()).keys())[0]]
    if blank_cols_no > (0.40*total_cols):
        print(col, "40%")
        pct40_blank_cols.append(col)
    if blank_cols_no > (0.50*total_cols):
        print(col, "50%")
        pct50_blank_cols.append(col)
    if blank_cols_no > (0.60*total_cols):
        print(col, "60%")
        pct60_blank_cols.append(col)
    if blank_cols_no > (0.70*total_cols):
        print(col, "70%")
        pct70_blank_cols.append(col)
    if blank_cols_no > (0.80*total_cols):
        print(col, "80%")
        pct80_blank_cols.append(col)
    if blank_cols_no > (0.90*total_cols):
        print(col, "90%")
        pct90_blank_cols.append(col)
```

```python
import pickle

with open('pct40_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct40_blank_cols, file)
with open('pct50_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct50_blank_cols, file)
with open('pct60_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct60_blank_cols, file)
with open('pct70_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct70_blank_cols, file)
with open('pct80_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct80_blank_cols, file)
with open('pct90_blank_cols.pkl', 'wb') as file:
    pickle.dump(pct90_blank_cols, file)
```

```python
useless_cols = pct50_blank_cols + ['Id', 'OrgId', 'IncidentId', 'AlertId']
useful_cols = [col for col in cols if col not in useless_cols]

import pandas as pd

# Define the path to the input and output files
input_file = "Data/GUIDE_train.csv"
output_file = "Data/GUIDE_train_cleaned.csv"

# Initialize a counter for the first chunk
chunk_counter = 0

# Process the CSV file in chunks
for chunk in pd.read_csv(input_file, chunksize=1000000, usecols=useful_cols):
    # Clean the chunk
    chunk.drop_duplicates(inplace=True)
    chunk = chunk.dropna(thresh=chunk.shape[1] - 20 + 1)
    chunk = chunk.dropna(subset=['IncidentGrade'])
    
    # If it's the first chunk, write the header, otherwise append without header
    if chunk_counter == 0:
        chunk.to_csv(output_file, index=False, mode='w')
    else:
        chunk.to_csv(output_file, index=False, mode='a', header=False)
    
    # Increment the chunk counter
    chunk_counter += 1
    print(chunk_counter)
```
